```python
import numpy as np

arr = np.array([1, 2, 3, 6, 4, 5])
reversed_arr = arr[::-1]
print(reversed_arr)
```

```python
array1 = np.array([[1, 2, 3], [2, 4, 5], [1, 2, 3]])

# Method 1: Using flatten()
flattened_array1 = array1.flatten()

# Method 2: Using ravel()
flattened_array2 = array1.ravel()

print(flattened_array1)
print(flattened_array2)
```

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[1, 2], [3, 4]])

# Check if arrays are equal
are_equal = np.array_equal(arr1, arr2)
print(are_equal)
```

```python
x = np.array([1,2,3,4,5,1,2,1,1,1])
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])

# Find the most frequent value in x
most_freq_value_x = np.bincount(x).argmax()
indices_x = np.where(x == most_freq_value_x)[0]

# Find the most frequent value in y
most_freq_value_y = np.bincount(y).argmax()
indices_y = np.where(y == most_freq_value_y)[0]

print("Most frequent value in x:", most_freq_value_x, "at indices", indices_x)
print("Most frequent value in y:", most_freq_value_y, "at indices", indices_y)
```

```python
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

# Sum of all elements
total_sum = np.sum(gfg)

# Sum row-wise
row_sum = np.sum(gfg, axis=1)

# Sum column-wise
column_sum = np.sum(gfg, axis=0)

print(total_sum, row_sum, column_sum)
```

```python
n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]])

# Sum of diagonal elements
diag_sum = np.trace(n_array)

# Eigenvalues and Eigenvectors
eigen_values, eigen_vectors = np.linalg.eig(n_array)

# Inverse of matrix
inverse = np.linalg.inv(n_array)

# Determinant of matrix
determinant = np.linalg.det(n_array)

print(diag_sum, eigen_values, eigen_vectors, inverse, determinant)
```

```python
p = np.array([[1, 2], [2, 3]])
q = np.array([[4, 5], [6, 7]])

# Multiply matrices
matrix_product = np.dot(p, q)

# Covariance
covariance = np.cov(p.flatten(), q.flatten())

print(matrix_product, covariance)
```

```python
x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])

# Inner product
inner_product = np.inner(x, y)

# Outer product
outer_product = np.outer(x, y)

# Cartesian product (flattened outer product)
cartesian_product = np.array([(a, b) for a in x.flatten() for b in y.flatten()])

print(inner_product, outer_product, cartesian_product)
```

```python
array = np.array([[1, -2, 3],[-4, 5, -6]])

# Element-wise absolute value
abs_array = np.abs(array)

# Percentiles
percentiles_flattened = np.percentile(array.flatten(), [25, 50, 75])
percentiles_column = np.percentile(array, [25, 50, 75], axis=0)
percentiles_row = np.percentile(array, [25, 50, 75], axis=1)

# Mean, Median, Standard Deviation
mean_flattened = np.mean(array.flatten())
median_flattened = np.median(array.flatten())
std_flattened = np.std(array.flatten())

mean_column = np.mean(array, axis=0)
median_column = np.median(array, axis=0)
std_column = np.std(array, axis=0)

mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)

print(abs_array, percentiles_flattened, mean_flattened, median_flattened, std_flattened)
```

```python
a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0])

floor_values = np.floor(a)
ceil_values = np.ceil(a)
truncated_values = np.trunc(a)
rounded_values = np.round(a)

print(floor_values, ceil_values, truncated_values, rounded_values)
```

```python
array = np.array([10, 52, 62, 16, 16, 54, 453])

# Sorted array
sorted_array = np.sort(array)

# Indices of sorted array
sorted_indices = np.argsort(array)

# 4 smallest elements
smallest_elements = np.partition(array, 4)[:4]

# 5 largest elements
largest_elements = np.partition(array, -5)[-5:]

print(sorted_array, sorted_indices, smallest_elements, largest_elements)
```

```python
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

# Integer elements only
integer_elements = array[array == array.astype(int)]

# Float elements only
float_elements = array[array != array.astype(int)]

print(integer_elements, float_elements)
```

```python
from PIL import Image
import numpy as np

def img_to_array(path):
    img = Image.open(path)
    img_array = np.array(img)

    if img.mode == 'RGB':
        np.savetxt('image_rgb.txt', img_array.reshape(-1, 3), fmt='%d')
    elif img.mode == 'L':
        np.savetxt('image_grey.txt', img_array, fmt='%d')

# Call the function with the path to an image
img_to_array('/content/338137417_172878452252056_8051241462967476920_n.jpg')
```
