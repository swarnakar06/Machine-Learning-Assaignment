```python
import numpy as np
# df['TotalChildren'] = df.replace(0,np.nan)
df['TotalChildren'] = df['TotalChildren'].replace(0, np.nan)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
df['TotalChildren'] = imputer.fit_transform(df[['TotalChildren']])
df['TotalChildren'].isna().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['YearlyIncome'] = scaler.fit_transform(df['YearlyIncome'].values.reshape(-1, 1))
df['YearlyIncome'].head()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder  = OneHotEncoder(sparse_output=False)
df['Gender'] = encoder.fit_transform(df[['Gender']])
df.head()
df['Gender'].head()
```

```python
df_numeric = df[['YearlyIncome','NumberCarsOwned']]
df_numeric.head()
```

```python
df_nominal = df[['Gender','MaritalStatus']]
df_nominal.head()
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, KBinsDiscretizer
from sklearn.pipeline import Pipeline

# Discretize and standardize numeric attributes
numeric_transformer = Pipeline([
    ('discretizer', KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')),
    ('scaler', StandardScaler())
])

# One-hot encode nominal attributes
nominal_transformer = OneHotEncoder(sparse_output=False)

```

```python
df_numeric = numeric_transformer.fit_transform(df_numeric)
df_nominal = nominal_transformer.fit_transform(df_nominal)
```

```python
df_numeric = pd.DataFrame(df_numeric)
df_nominal = pd.DataFrame(df_nominal)
```

```python
df_numeric.head()
```

```python
df_nominal.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# Calculate Hamming Distance
nominal_similarity = cosine_similarity(df_nominal)
print("Nominal Similarity (Hamming Distance):\n", nominal_similarity)

numeric_similarity = cosine_similarity(df_numeric)
print("\nNumeric Similarity (Euclidean Distance):\n", numeric_similarity)
```
